- Para esse projeto, escolhi utilizar o dataset ``Traditional Food around the world`` presente no site [Kaggle](https://www.kaggle.com/abhijeetbhilare/world-cuisines). O dataset é formado por diversas imagens que demonstram pratos culinários tradicionais de diversas regiões do mundo, como por exemplo ``American``, ``European`` e ``Indian``. O objetivo desse projeto é dado um prato determinar o país de origem do mesmo. Além dessa classificação, usaremos explicabilidade para tentar entender o motivo do modelo ter aprendido que um prato X pertence à região Y.<br><br>

- Para a leitura das imagens e armazenamento utilizaremos as bibliotecas OpenCV e glob.
- Para a manipulação dos dados utilizaremos as bibliotecas NumPy.
- Para a exibição de gráficos e resultados utilizaremos a biblioteca Matplotlib.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from glob import glob
```

# Leitura dos dados e pré-processamento
## 1. Definindo o caminho para os dados

```python
paths   = dict()
classes = ["American", "Chinese", "European", "Indian", "Japanese", "Korean"]

for _class in classes:
    paths[_class] = "Dishes/" + _class + "/"

print(paths)
```

## 2. Guardando os dados

```python
Dishes = dict()

for _class in classes:
    Dishes[_class] = [cv2.cvtColor(cv2.imread(img), cv2.COLOR_BGR2RGB) for img in glob(paths[_class] + "*.jpg")]
    print(_class + " has %d images" % len(Dishes[_class]))
```

- Como visto acima, os dados são bem distribuídos ao longo das classes, exceto para a classe Chinese que possui $\approx 1.67$ dados a mais que a classe American.

## 3. Exibição de algumas imagens

```python
fig, axs = plt.subplots(2, 3)
    
axs[0, 0].imshow(Dishes["American"][42])
axs[0, 0].set_title("American Dish")

axs[0, 1].imshow(Dishes["Chinese"][42])
axs[0, 1].set_title("Chinese Dish")

axs[0, 2].imshow(Dishes["European"][42])
axs[0, 2].set_title("European Dish")

axs[1, 0].imshow(Dishes["Indian"][42])
axs[1, 0].set_title("Indian Dish")

axs[1, 1].imshow(Dishes["Japanese"][42])
axs[1, 1].set_title("Japanese Dish")

axs[1, 2].imshow(Dishes["Korean"][42])
axs[1, 2].set_title("Korean Dish")

for ax in axs.flat:
    ax.label_outer()

print("Shape of each image:", Dishes["American"][42].shape)
plt.show()
```
